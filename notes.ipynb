<h1>Django Notes</h1>

To create a django project ---> django-admin startproject project name

To create a main project where the website is ---> python manage.py startapp app_name

To run prject locally ---> python manage.py runserver

<h1>MVT (Model View Tempelate) Django Archi</h1>

MVT is a software design pattern. 

Model ---> Information to create a db. Like, primary key, fields etc

View ---> Executes the business logic and interact with a model to carry data and renders a template.

View will be created only one.

Template --> Template is simply html, css, js but used django templatic engine.
